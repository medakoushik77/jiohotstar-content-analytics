```python
missing_values = df_clean.isnull().sum()

missing_percentage = (missing_values / len(df_clean)) * 100

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Percentage (%)": missing_percentage.round(2)
})

missing_summary.sort_values(by="Percentage (%)", ascending=False)
```

```python
df_clean["added_date"] = pd.to_datetime(df_clean["added_date"])
```

```python
df_clean.info()
```

```python
categorical_columns = [
    "title",
    "content_type",
    "genre",
    "sub_genre",
    "language",
    "country",
    "age_rating",
    "director",
    "production_company",
    "is_original"
]

for column in categorical_columns:
    df_clean[column] = df_clean[column].fillna(df_clean[column].mode()[0])
```

```python
df_clean.isnull().sum()
```

```python
numerical_columns = [
    "release_year",
    "cast_count",
    "imdb_rating",
    "jiohotstar_rating",
    "watch_count_millions",
    "completion_rate",
    "awards_count"
]

for column in numerical_columns:
    df_clean[column] = df_clean[column].fillna(df_clean[column].median())
```

```python
df_clean.isnull().sum()
```

# Phase 3 – Data Cleaning Summary

## Key Decisions

- Created a working copy of the dataset (`df_clean`) to preserve the original data.
- Converted `added_date` to `datetime` format.
- Filled missing values in categorical columns using the **mode** (most frequent value).
- Filled missing values in numerical columns using the **median** to reduce the effect of outliers.
- Preserved missing values in `duration_minutes`, `seasons`, and `episodes` because they are logically valid for Movies and TV Shows.
- Left missing values in `added_date` unchanged because there was insufficient information to infer the correct dates.

## Result

The dataset is now clean, consistent, and ready for Exploratory Data Analysis (EDA).

## Analysis 10: Correlation Analysis

### Business Question
Which numerical features are most strongly related to each other?

### Objective
Analyze the correlation between ratings, watch counts, completion rates, awards, duration, and other numerical variables to uncover meaningful business relationships.

```python


correlation = df_clean[
    [
        "release_year",
        "duration_minutes",
        "cast_count",
        "imdb_rating",
        "jiohotstar_rating",
        "watch_count_millions",
        "completion_rate",
        "awards_count"
    ]
].corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title(
    "Analysis 10: Correlation Heatmap",
    fontsize=14,
    fontweight="bold"
)

plt.savefig(
    "../images/correlation_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

### Business Insight

- The heatmap summarizes the relationships between all numerical variables.
- Positive correlation values indicate that two variables tend to increase together, while negative values indicate an inverse relationship.
- Most variables in this dataset show weak to moderate correlations, suggesting that multiple factors influence audience engagement rather than a single metric.
- Correlation analysis helps identify patterns that can guide future content strategy and deeper predictive analysis.
